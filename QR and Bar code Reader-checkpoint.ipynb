```python
#for video capture
cap = cv2.VideoCapture('qrcode.mp4')
cap.set(3, 640)
cap.set(4, 480)

while(True):
    ret, frame = cap.read()
    if ret == True:
        for barcode in decode(frame):
            mydata = barcode.data.decode('utf-8')
            #draw a poly 
            pts = np.array([barcode.polygon], np.int32)
            pts = pts.reshape((-1,1,2))
            #print(pts)
            cv2.polylines(frame, [pts], True, (0,255,0),5)
            pts2 = barcode.rect
            cv2.putText(frame, mydata, (pts2[0], pts2[1]), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 3)
    
        cv2.imshow('frame',frame)
        if cv2.waitKey(1) & 0xFF == 27:
            break
            
cap.release()
cv2.destroyAllWindows()
```

```python
#Authentication using Barcode

#for video capture
cap = cv2.VideoCapture('qrcode.mp4')
cap.set(3, 640)
cap.set(4, 480)

with open('qrtext.txt') as f:
    mydatalist = f.read().splitlines()

while(True):
    ret, frame = cap.read()
    if ret == True:
        for barcode in decode(frame):
            mydata = barcode.data.decode('utf-8')
            print(mydata)
            
            if mydata in mydatalist:
                #draw a poly 
                pts = np.array([barcode.polygon], np.int32)
                pts = pts.reshape((-1,1,2))
                
                cv2.polylines(frame, [pts], True, (0,255,0),3)
                pts2 = barcode.rect
                cv2.putText(frame, 'Authorized', (pts2[0], pts2[1]), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 3)
            
            else:
                #draw a poly 
                pts = np.array([barcode.polygon], np.int32)
                pts = pts.reshape((-1,1,2))
                
                cv2.polylines(frame, [pts], True, (0,0,255),5)
                pts2 = barcode.rect
                cv2.putText(frame, 'Unauthorized', (pts2[0], pts2[1]), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 3)
    
        cv2.imshow('frame',frame)
        if cv2.waitKey(1) & 0xFF == 27:
            break
            
cap.release()
cv2.destroyAllWindows()
```
