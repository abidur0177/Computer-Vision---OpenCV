```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#smoothing means filter. There are verious linear filters
fig = plt.figure(figsize=(12,8))
img = cv2.imread('salt.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#Homogeneous filter the simplest filter. It measure the mean of the kernel neighbors
kernel = np.ones((5,5), np.float32)/25 #the homogeneous kernel
dst = cv2.filter2D(img, -1,kernel) #this is the normal filter method

#images can be filtered with various LPF and HPF. LPS remove noise and blur images. HPF finds edges of the images.

#Blur = Averaging Filter
blur = cv2.blur(img,(5,5))

#Gaussian Blur
gblur = cv2.GaussianBlur(img,(5,5),0)

#Median Filter, replace each pixel by the median of its neighbors pixel. Great when dealing with salt and pepper noise
mblur = cv2.medianBlur(img,3)

#Bilateral filter = Preserve the edges even though it smooth the image
bilateral = cv2.bilateralFilter(img, 7, 75,75)


titles=['Image','2D Conv','Blur','Gaussian Blur','Median Blur','Bilateral']
images = [img,dst,blur,gblur,mblur, bilateral]

for i in range(6):
    plt.subplot(2,3,i+1), plt.imshow(images[i],'gray')
    plt.title(titles[i],color='red')
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))
img = cv2.imread('lena.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel = np.ones((5,5), np.float32)/25

d2 = cv2.filter2D(img, -1, kernel)
blur = cv2.blur(img, (5,5))
gblur = cv2.GaussianBlur(img, (5,5), 0)
mblur = cv2.medianBlur(img, 3)
bilateral = cv2.bilateralFilter(img, 5, 75,75)

images = [img, d2, blur,gblur,mblur, bilateral]
titles = ['Original', 'Filter 2D', 'Blur', 'Gaussian Blur','Median Blur','Bilateral']

for i in range(len(images)):
    plt.subplot(2,3, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
