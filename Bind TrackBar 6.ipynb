```python
import cv2
import numpy as np
```

```python
def nothing(x): #x is the position of the trackbar
    print(x) #when the trackbar value changes this func will be called

img = np.zeros((300,512,3), np.uint8)
cv.namedWindow('Image')

#now bind a switch trackbar
switch = '0: OFF \n 1: ON'
    

#create a trackbar. we can create multiple trackbar for an image
cv.createTrackbar('B', 'Image',0, 255, nothing)
cv.createTrackbar('G', 'Image',0, 255, nothing)
cv.createTrackbar('R', 'Image',0, 255, nothing)
cv.createTrackbar(switch,'Image',0, 1, nothing)


while(True):
        
    #get the trackbar position
    b = cv.getTrackbarPos('B','Image')
    g = cv.getTrackbarPos('G','Image')
    r = cv.getTrackbarPos('R','Image')
    s = cv.getTrackbarPos(switch,'Image')
    
    if s==0:
        img[:] = 0
    else:
        img[:] = [b,g,r]
    
    cv.imshow('Image', img)
    k = cv.waitKey(1) & 0xFF
    if k ==27:
        break
    
cv.destroyAllWindows()
```
