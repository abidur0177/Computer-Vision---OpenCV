### Face detection in an image

```python
image = face_recognition.load_image_file('face_detection.jpg')
face_locations = face_recognition.face_locations(image)
images = []

print("I have detected {} faces in this picture".format(len(face_locations)))

for face_location in face_locations:
    top, right, bottom, left = face_location
    print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(top, left, bottom, right))
    cv2.rectangle(image, (top,left), (bottom,right), (25,255,2), 2)
    face_image = image[top:bottom, left:right]
    images.append(face_image)
    

for i in range(len(images)):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
cv2.imshow('Image', cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

### Face detection using deep learning

```python
image = face_recognition.load_image_file('face_detection.jpg')
face_locations = face_recognition.face_locations(image, number_of_times_to_upsample=0, model='cnn')

print("I found {} face(s) in this photograph.".format(len(face_locations)))
for face_location in face_locations:
    top, right, bottom, left = face_location
    print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(top, left, bottom, right))
    face_image = image[top:bottom, left:right]
    plt.imshow(face_image)
plt.show()
```

### Detect facial landmarks

```python
from PIL import Image, ImageDraw

image = face_recognition.load_image_file('two_people.jpg')

face_landmarks = face_recognition.face_landmarks(image)
print("I found {}face(s) in the photograph".format(len(face_landmarks)))

# Create a PIL imagedraw object so we can draw on the picture
pil_image = Image.fromarray(image)
d = ImageDraw.Draw(pil_image)

for face_landmark in face_landmarks:
    for facial_feature in face_landmark.keys():
        print("The {} in this face has the following points: {}".format(facial_feature, face_landmark[facial_feature]))
        print('\n')
        
    for facial_feature in face_landmark.keys():
        d.line(face_landmark[facial_feature], width=5)

pil_image.show()
```
