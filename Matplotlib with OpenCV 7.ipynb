```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread('butterfly.jpg')

cv2.imshow('Image', img) #read as BGR format
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img) #read as RGB format. So it will show accurate when converted to RGB format
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Showing various graph on matplotlib

```python
fig,ax = plt.subplots(3,2, facecolor='y')

#thresholding is used to seperate a object from its bg. compare each pixel of an image to a predefined Th value
img = cv2.imread('gradient.jpg')
#we have to define two variable. one is ret and the other is threshold value
_,th_bin = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
_,th_bin_inv = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
_,th_trun = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
_,th_tozero = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
_,th_tozero_inv = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)

ax[0,0].imshow(th_bin)
ax[0,1].imshow(th_bin_inv)
ax[1,0].imshow(th_trun)
ax[1,1].imshow(th_tozero)
ax[2,0].imshow(th_tozero_inv)

plt.show()
```

```python
#thresholding is used to seperate a object from its bg. compare each pixel of an image to a predefined Th value
img = cv2.imread('gradient.jpg')
#we have to define two variable. one is ret and the other is threshold value
_,th_bin = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
_,th_bin_inv = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
_,th_trun = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
_,th_tozero = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
_,th_tozero_inv = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)

images =[img,th_bin,th_bin_inv, th_trun, th_tozero, th_tozero_inv]
titles = ['Original', 'Binary', 'Binary Inverse', 'Truncate', 'To Zero', 'To Zero Inverse']
for i in range(1,6+1):
    plt.subplot(2,3,i)
    plt.imshow(images[i-1])
    plt.title(titles[i-1], color='red')
    plt.show()

plt.show()
```
