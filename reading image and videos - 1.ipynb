### Reading Video with opencv

```python
cap = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*'XVID')

out = cv2.VideoWriter('camera.avi', fourcc, 30, (640,480))
print(cap.isOpened())
while(cap.isOpened()):
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    print(cap.get(cv2.CAP_PROP_AUTO_EXPOSURE))
    
    if ret is True:
        out.write(frame)
        cv2.imshow('frame', frame)

        if cv2.waitKey(1) & 0xFF ==ord('q'):
            break
cap.release()
out.release()
cv2.destroyAllWindows()
```

### Draw geometric shapes in opencv

```python
img = cv2.imread('starry_night.jpg',1)

img = cv2.line(img, (50,100), (255,255), (25,233,211), 5) # normal line
img = cv2.arrowedLine(img, (100,200), (400,400), (148,83,111), 5) # arrowed line
img = cv2.rectangle(img, (400,40),(500,200), (48,93,11), 10) # rectangle (the 5th argument is thickness. if given -1 then fill the rectangle)
img = cv2.circle(img, (450,450), 100, (18,143,181), -1) # circle
img = cv2.ellipse(img, (300,300),(100,50),45,100,270,(18,13,81),5,cv2.LINE_AA)
#font for text
font = cv2.FONT_HERSHEY_TRIPLEX
img = cv2.putText(img,'Hoga mara sara', (110,450),font, 2, (188,143,181), 5, cv2.LINE_AA) #text (8th argument is the line type)

cv2.imshow('Starry night', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Create Image using numpy zeros method

```python
img = np.zeros([600, 700,3], np.uint8) #[height, width, 3]

img = cv2.line(img, (50,100), (255,255), (25,233,211), 5) # normal line
img = cv2.arrowedLine(img, (100,200), (400,400), (148,83,111), 5) # arrowed line
img = cv2.rectangle(img, (400,40),(500,200), (48,93,11), 10) # rectangle (the 5th argument is thickness. if given -1 then fill the rectangle)
img = cv2.circle(img, (450,450), 100, (18,143,181), -1) # circle
#font for text
font = cv2.FONT_HERSHEY_TRIPLEX
img = cv2.putText(img,'Hoga mara sara', (110,450),font, 2, (188,143,181), 5, cv2.LINE_AA) #text (8th argument is the line type)

cv2.imshow('Starry night', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#video read
cap = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*'XVID')

out = cv2.VideoWriter('camera.mp4', fourcc, 30, (700,480))

while(cap.isOpened()):
    ret, frame = cap.read()
    
    if ret == True:
        cv2.imshow('Frame', frame)
        out.write(frame)
        if cv2.waitKey(1) & 0xFF == 27:
            break
cap.release()
cv2.destroyAllWindows()
```
