```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('shape.jpg')
#convert to gray image
imgg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#get the threshold
_, thresh = cv2.threshold(imgg, 240,255,cv2.THRESH_BINARY)
#find the contours
contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)

for contour in contours:
    approx = cv2.approxPolyDP(contour, 0.01*cv2.arcLength(contour, True), True) #approxPolyDP find potential polygon
#in image. contour in the src, arclength measures the arclength of the contour. True means closed curve. 
#This return the number of polygonal curve
    cv2.drawContours(img, [approx], 0, (0,0,0), 2)
    print(approx)
    
    x = approx.ravel()[0] #get the co-ordinate of the shape where to write
    print(x)
    y = approx.ravel()[1]
    print(y)
    
    if len(approx) == 3:
        cv2.putText(img, 'Triangle', (x,y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0,0,134), 2)
    elif len(approx) == 4:
        (x,y,w,h) = cv2.boundingRect(approx)
        aspect_ratio = float(w/h)
        if aspect_ratio>=9.5 and aspect_ratio<=1:
            cv2.putText(img, 'Square', (x,y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0,0,134), 2)
        else:
            cv2.putText(img, 'Rectangle', (x,y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0,0,134), 2)
    elif len(approx) == 5:
        cv2.putText(img, 'Pentagon', (x,y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0,0,134), 2)
    elif len(approx) == 10:
        cv2.putText(img, 'Star', (x,y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0,0,134), 2)
    else:
        cv2.putText(img, 'Circle', (x,y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0,0,134), 2)
    
    
cv2.imshow('Shape',img)
cv2.imshow('Shape Gray',thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('shape.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_,th = cv2.threshold(img_gray, 20, 255, cv2.THRESH_BINARY)
contours,_ = cv2.findContours(th, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)

for contour in contours:
    approx = cv2.approxPolyDP(contour, 0.001*cv2.arcLength(contour, True), True)
    cv2.drawContours(img, [approx], -1, (147,222,21), 2)
    
    x = approx.ravel()[0]
    y = approx.ravel()[1]
    font = cv2.FONT_HERSHEY_COMPLEX
    
    if len(approx) == 3:
        cv2.putText(img, 'Triangle', (x,y), font, 2, (0,0,255), 2)
    elif len(approx) == 5:
        cv2.putText(img, 'Pentagon', (x,y), font, 2, (0,0,255), 2)
    elif len(approx) == 4:
        (x,y,w,h) = cv2.boundingRect(approx)
        ratio = float(w/h)
        if ratio >=9.51 and ratio <=1:
            cv2.putText(img, 'Square', (x,y), font, 2, (0,0,255), 2)
        else:
            cv2.putText(img, 'Rectangle', (x,y), font, 2, (0,0,255), 2)
    else:
        cv2.putText(img, 'Circle', (x,y), font, 2, (0,0,255), 2)
cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
    
```
