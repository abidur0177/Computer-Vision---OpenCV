```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = np.zeros((200,200,3), np.uint8)
cv2.rectangle(img, (10,20), (150,150), (24,52,221), -1)
cv2.rectangle(img, (0,20), (100,150), (123,132,21), -1)
img = cv2.imread('lena.jpg',0)

cv2.imshow('Image',img)


plt.hist(img.ravel(), 256, [0,256]) #histogram with matplotlib, first ravel of image, 2nd max num of pixel value, 3rd
#range 0 to 256
plt.xticks(color='red')
plt.yticks(color='r')
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = np.zeros((200,200), np.uint8)
cv2.rectangle(img, (10,20), (150,150), (234,52,221), -1)
cv2.rectangle(img, (0,20), (100,150), (123,2,21), -1)
img = cv2.imread('lena.jpg')
b,g,r = cv2.split(img)

cv2.imshow('Image',img)
cv2.imshow('B',b)
cv2.imshow('G',g)
cv2.imshow('R',r)


plt.hist(img.ravel(), 256, [0,256], label='Image') #histogram with matplotlib, first ravel of image, 
#2nd max num of pixel value, 3rd range 0 to 256
plt.hist(b.ravel(), 256, [0,256], label='Blue')
plt.hist(g.ravel(), 256, [0,256], label='Green')
plt.hist(r.ravel(), 256, [0,256], label='Red')

plt.xticks(color='red')
plt.yticks(color='r')
plt.legend()
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#now plot histogram using cv2
hist = cv2.calcHist([img],[1],None,[256],[0,256])
plt.plot(hist)
```
