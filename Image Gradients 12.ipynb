```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Image gradient is an directional change in the intensity or color in an image. Is used to 
#find the edges inside an image
fig = plt.figure(figsize=(16,8))
img = cv2.imread('sudoku.png',0)

#Laplacial gradient
lap = cv2.Laplacian(img, cv2.CV_64F,ksize=3) #cv_64F in dtype which support negative value 
#when convert from BGR to GRAY
#now take the absolute value of Lap Transformation and convert it back to 
#the unsigned 8bit integer which is suitable for output
lap = np.uint8(np.absolute(lap))

#Sobel_X
sobel_x = cv2.Sobel(img,cv2.CV_64F, 1,0) #3rd arg is dx, when been chose 1, means x_sobel is running
sobel_x = np.uint8(np.absolute(sobel_x))

#Soble_y
sobel_y = cv2.Sobel(img,cv2.CV_64F, 0,1)
sobel_y = np.uint8(np.absolute(sobel_y))

#Soble_xy
sobel_combined = cv2.bitwise_or(sobel_x,sobel_y)

titles = ['Image','Laplacian','Sobel X','Sobel Y','Sobel Combined']
images = [img,lap,sobel_x, sobel_y,sobel_combined]

for i in range(5):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i], color='red')
    plt.xticks([])
    plt.yticks([])

plt.show()
```
